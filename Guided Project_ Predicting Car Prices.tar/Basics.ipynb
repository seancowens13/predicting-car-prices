```python
import pandas as pd

cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
imports = pd.read_csv('imports-85.data', names=cols)
imports
```

```python
import numpy as np

imports = imports.replace('?', np.nan)
imports['num-of-doors'] = imports['num-of-doors'].replace('four', 4)
imports['num-of-doors'] = imports['num-of-doors'].replace('two', 2)
imports.head()
```

```python
continuous_cols = ['normalized-losses', 'num-of-doors', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

for col in continuous_cols:
    imports[col] = imports[col].astype('float')
imports.head()
```

```python
imports.info()
```

```python
imports.isnull().sum()
```

```python
imports = imports.dropna(subset=['price'])
imports.isnull().sum()
```

```python
imports = imports.fillna(imports.mean())
imports.isnull().sum()
```

```python
price_col = imports['price']

cars = imports[continuous_cols]
cars = (cars - cars.min()) / (cars.max() - cars.min())
cars['price'] = price_col
```

```python
cars.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col, target_col, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)
    
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    last_train_row = int(len(rand_df) / 2)
    
    train_df = rand_df[:last_train_row]
    test_df = rand_df[last_train_row:]
    
    knn.fit(train_df[[train_col]], train_df[target_col])
    
    predictions = knn.predict(test_df[[train_col]])
    
    mse = mean_squared_error(test_df[target_col], predictions)
    rmse = np.sqrt(mse)
    return rmse

rmses = {}
train_cols = cars.columns.drop('price')

for col in train_cols:
    rmse_val = knn_train_test(col, 'price', cars)
    rmses[col] = rmse_val
    
rmse_results = pd.Series(rmses)
rmse_results.sort_values()
```

```python
def knn_train_test(train_col, target_col, df):
    np.random.seed(1)
    
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    last_train_row = int(len(rand_df) / 2)
    
    train_df = rand_df[:last_train_row]
    test_df = rand_df[last_train_row:]
    
    k_vals = [1,3,5,7,9]
    k_rmses = {}
    
    for k in k_vals:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[train_col]], train_df[target_col])
        
        predictions = knn.predict(test_df[[train_col]])
        
        mse = mean_squared_error(test_df[target_col], predictions)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses

k_rmses_results = {}

train_cols = cars.columns.drop('price')
for col in train_cols:
    rmse_val = knn_train_test(col, 'price', cars)
    k_rmses_results[col] = rmse_val
    
k_rmses_results
```
